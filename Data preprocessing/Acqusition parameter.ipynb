```python
import os
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```

```python
CPTAC_LAUD_root = 'D:\\CPTAC_LAUD\\image'
```

```python
CPTAC_LAUD_data = os.listdir(CPTAC_LAUD_root)
```

```python
count_size = np.array([]).reshape(0, 4)
count_spacing = np.array([]).reshape(0, 4)
```

```python
for data in CPTAC_LAUD_data:
    if 'roi' not in data:
        num, _, _ = data.split('.')
        str_num = np.array([num])
        
        img_root = os.path.join(CPTAC_LAUD_root, data)
        img = sitk.ReadImage(img_root)
        size = np.array(img.GetSize())
        spacing = np.array(img.GetSpacing())
        
        num_size = np.append(str_num, size).reshape(1, 4)
        count_size = np.append(count_size, num_size, axis = 0)
        num_spacing = np.append(str_num, spacing).reshape(1, 4)
        count_spacing = np.append(count_spacing, num_spacing, axis = 0)
```

```python
CPTAC_LAUD_spacing = pd.DataFrame(count_spacing, columns = ['num', 'collimation_x', 'collimation_y', 'reconstruction'])
```

```python
CPTAC_LAUD_spacing
```
